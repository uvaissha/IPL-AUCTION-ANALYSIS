```python
plt.figure(figsize=(12,6))
sns.histplot(ipl['base price (in lacs)'],bins=30,color='blue',label='base price')

plt.title('Distribution of Base Price ')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="player style",data=ipl,palette="Set2")
plt.title("Player Style Distribution")
plt.xlabel("Player Style")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="franchise",data=ipl,palette="viridis")
plt.title("franchise Distribution")
plt.xlabel("franchise")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="nationality",data=ipl,palette="coolwarm")
plt.xticks(rotation=45)
plt.title("nationality Distribution")
plt.xlabel("nationality")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x="base price (in lacs)",y="final price (in lacs)",data=ipl,hue="status",color="red")
plt.title("Base Price vs Final Price")
plt.xlabel("Base Price")
plt.ylabel("Final Price")
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=ipl[['base price (in lacs)', 'final price (in lacs)']])
plt.title("Base Price vs Final Price")
plt.xlabel("Base Price")
plt.ylabel("Final Price")
plt.show()
```

```python

Q1_base = ipl['base price (in lacs)'].quantile(0.25)
Q3_base = ipl['base price (in lacs)'].quantile(0.75)
IQR_base = Q3_base - Q1_base


lower_bound_base = Q1_base - 1.5 * IQR_base
upper_bound_base = Q3_base + 1.5 * IQR_base

outliers_base = ipl[(ipl['base price (in lacs)'] < lower_bound_base) | (ipl['base price (in lacs)'] > upper_bound_base)]

print("Outliers in 'base price (in lacs)':")
print(outliers_base)

Q1_final = ipl['final price (in lacs)'].quantile(0.25)
Q3_final = ipl['final price (in lacs)'].quantile(0.75)
IQR_final = Q3_final - Q1_final


lower_bound_final = Q1_final - 1.5 * IQR_final
upper_bound_final = Q3_final + 1.5 * IQR_final

outliers_final = ipl[(ipl['final price (in lacs)'] < lower_bound_final) | (ipl['final price (in lacs)'] > upper_bound_final)]

print("Outliers in 'final price (in lacs)':")
print(outliers_final)
```
